# [실습] 비행기 연착 추측 분류
- 비행정보, 출발 공항 정보, 도착 공항 정보 병합 (Airport ID)
- target : ArrDel15

```python
import pandas as pd
import numpy as np

flight_time = pd.read_csv('./dataset/Flight on-time performance.csv')
print(flight_time.shape)
display(flight_time.head())
```

```python
ori_airport = pd.read_csv('./dataset/Airport Codes Dataset.csv')
print(ori_airport.shape)
display(ori_airport.head())
```

```python
ori_airport.columns = ['airport_id', 'ori_city', 'ori_state', 'ori_airport']
display(ori_airport.head())
```

```python
dest_airport = pd.read_csv('./dataset/Airport Codes Dataset.csv')

dest_airport.columns = ['airport_id', 'dest_city', 'dest_state', 'dest_airport']
display(dest_airport.head())
```

```python
# 데이터 병합

df = pd.merge(flight_time, ori_airport, left_on='OriginAirportID', right_on='airport_id')
df = df.drop('airport_id', axis=1)
print(df.shape)
display(df.head())
```

```python
df = pd.merge(df, dest_airport, left_on='DestAirportID', right_on='airport_id')
print(df.shape)
display(df.head())
```

```python
df = df.drop('airport_id', axis=1)
print(df.shape)
display(df.head())
```

```python
# 학습 컬럼 선택
flight = df[['DayOfWeek', 'Carrier', 'DepTimeBlk', 'DepDelay', 'DepDel15', 'ArrTimeBlk', 'ArrDel15', 'ori_city', 'ori_state', 'ori_airport', 'dest_city', 'dest_state', 'dest_airport', 'ArrDelay']]
display(flight.head())
```

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
# 결측치 삭제
flight.isna().sum()
flight.dropna(inplace=True)
```

```python
# 카테고리형으로 타입 변경
flight['ArrDel15'] = flight['ArrDel15'].astype('category')
flight.info()
```

```python
# 라벨 인코딩
from sklearn.preprocessing import LabelEncoder

for col in flight.columns:
    if flight[col].dtype == 'O':
        flight[col] = LabelEncoder().fit_transform(flight[col])

flight
```

```python
flight['Carrier'].dtype
```

```python
flight['ArrDel15'].unique()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
x = flight.drop('ArrDel15', axis=1)
y = flight['ArrDel15']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

## 로지스틱 회귀모형

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_hat = model.predict(x_test)

print(f'정확도: {accuracy_score(y_test, y_hat):.3f}')
print(f'정밀도: {precision_score(y_test, y_hat):.3f}')
```

```python
con_mat = confusion_matrix(y_test, y_hat)
con_mat
```

## Ensemble - Bagging

```python
from sklearn.ensemble import RandomForestClassifier

r_forest = RandomForestClassifier()
r_forest.fit(x_train, y_train)
```

```python
y_hat = r_forest.predict(x_test)

print(f'정확도: {accuracy_score(y_test, y_hat):.3f}')
print(f'정밀도: {precision_score(y_test, y_hat):.3f}')
```

```python
con_mat = confusion_matrix(y_test, y_hat)
con_mat
```

## Ensemble - Voting

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier

logistic = LogisticRegression(max_iter=4000) # 반복학습 1000번
knn = KNeighborsClassifier()

vo_c = VotingClassifier(estimators=[('LOGISTIC', logistic),('KNN', knn)], voting='hard')
```

```python
vo_c.fit(x_train, y_train)
y_hat = vo_c.predict(x_test)

print('실제값:', np.array(y_test[:7]))
print('예측값:', y_hat[:7])

print(f'정확도: {accuracy_score(y_test, y_hat):.3f}')
print(f'정밀도: {precision_score(y_test, y_hat):.3f}')
```

## 패키지 로딩

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

## 데이터 로딩

```python
df_airport = pd.read_csv('./dataset/Airport Codes Dataset.csv')
print(df_airport.shape)
display(df_airport.head())
```

```python
df_flight = pd.read_csv('./dataset/Flight on-time performance.csv')
print(df_flight.shape)
display(df_flight.head())
```

```python
df_flight['ArrDel15'].value_counts()
```

```python
# 결측치 확인
print(df_airport.isna().sum())
print(df_flight.isna().sum())
```

## 데이터 전처리

### 공항정보 데이터를 출발 공항정보로 수정

```python
df_ori_airport = df_airport.rename(columns={'city':'Ori-city', 'state':'Ori-state', 'name':'Ori-airport'})
df_ori_airport.head()
```

### 비행정보 데이터와 출발공항정보 데이터 병합

```python
merged_df1 = pd.merge(df_flight, df_ori_airport, left_on='OriginAirportID', right_on='airport_id', how='left')
print(merged_df1.shape)
display(merged_df1.head())
```

```python
# airport id 중복 컬럼 삭제
merged_df1.drop('airport_id', axis=1, inplace=True)
```

### 공항정보 데이터를 도착 공항정보로 수정

```python
df_dest_airport = df_airport.rename(columns={'city':'Dest-city', 'state':'Dest-state', 'name':'Dest-airport'})
df_dest_airport.head()
```

### 비행정보 데이터와 출발공항정보 데이터 병합

```python
merged_df2 = pd.merge(merged_df1, df_dest_airport, left_on='DestAirportID', right_on='airport_id', how='left')
print(merged_df2.shape)
display(merged_df2.head())
```

```python
# airport id 중복 컬럼 삭제
merged_df2.drop('airport_id', axis=1, inplace=True)
```

### 24개 컬럼 중 학습에 사용할 14개 컬럼 선택

```python
merged_df2.columns
```

```python
selected_col = ['DayOfWeek', 'Carrier','DepTimeBlk','DepDelay','DepDel15','ArrTimeBlk', 'ArrDelay','ArrDel15','Ori-city', 'Ori-state',
       'Ori-airport', 'Dest-city', 'Dest-state', 'Dest-airport']
df_flight_info = merged_df2[selected_col]
df_flight_info.head()
```

## 결측치 처리

```python
df_flight_info = df_flight_info.dropna()
```

```python
df_flight_info.isna().sum()
```

```python
df_flight_info.info()
```

### 명목형 데이터의 유니크 값 개수 확인

```python
obj_col = [ col for col in df_flight_info.columns if df_flight_info[col].dtype == 'object' ]
for col in obj_col:
    print(f'컬럼명: {col:<15s}, 유니크 값 개수:{df_flight_info[col].unique().size}')
```

### 라벨 인코딩

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
for col in [ col for col in df_flight_info.columns if df_flight_info[col].dtype == 'object' ]:
    df_flight_info[col] = encoder.fit_transform(df_flight_info[col])
```

```python
df_flight_info.info()
```

## 데이터 분할 및 모델 생성

```python
from sklearn.model_selection import train_test_split

x = df_flight_info.drop('ArrDel15', axis=1)
y = df_flight_info['ArrDel15']

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, stratify=y)
```

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
model = LogisticRegression(solver = 'sag', max_iter=600) # sag: 확률적 경사하강법
model.fit(x_train, y_train)
```

## 모델 평가

```python
from sklearn.metrics import accuracy_score, roc_auc_score

y_hat = model.predict(x_test)
print(f'정확도: {accuracy_score(y_test, y_hat):.3f}')
print(f'AUC: {roc_auc_score(y_test, model.predict_proba(x_test)[:,1])}')
```

## Feature Importance - Random forest classifier

```python
importance = { k:v for k, v in zip(x.columns, model.feature_importances_)}
df_importance = pd.DataFrame(pd.Series(importance), columns=['importance'])
# 딕셔너리의 키값이 인덱스로 가야하기 때문에 시리즈로 데이터 프레임 생성
df_importance.sort_values('importance', ascending=False)
```

## 회귀계수(가중치) 확인 - Logistic Regression

```python
coef = model.coef_.squeeze(axis=0)
odds_rate = np.exp(coef)

coef_df = pd.DataFrame({'가중치':coef, 'Odds비':odds_rate}, index=x.columns)
coef_df.sort_values('가중치', ascending=False)

# 회귀계수 해석
# 로지스틱 회귀계수에 지수변환(exp())을 적용하면 odds비가 나온다. 오즈비= 성공확률/실패확률 
# odds비가 크면 연착 확률이 높음
```

## 교차검증

```python
from sklearn.model_selection import cross_validate

scores = cross_validate(model, x, y, cv=5, scoring=['accuracy', 'roc_auc_ovr'])
for k,v in scores.items():
    print('Key:',k)
    print(f'평균값: {np.mean(v):.3f}')
    print('-'*30)
```
