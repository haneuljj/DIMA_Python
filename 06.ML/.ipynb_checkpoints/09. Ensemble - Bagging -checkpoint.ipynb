# Bagging - Regressor

## 보스톤 집값 예측

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import numpy as np
```

## 데이터 로드

```python
boston = pd.read_csv('./dataset/HousingData.csv')
boston.head()
boston.dropna(inplace=True)

x = boston.drop('MEDV', axis=1)
y = boston['MEDV']
```

## 학습/평가 데이터 분할 맟 스케일링

```python
scaled_x = StandardScaler().fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(scaled_x, y, test_size=0.3, random_state=10)
```

## 모델 생성 및 검증

```python
model = RandomForestRegressor(n_estimators=150, random_state=0) 
# n_estimator: 분류기(나무모형)을 몇개 만들 것인지
# 원데이터를 임의로 복원 추출하여 학습(bagging)
model.fit(x_train, y_train)
```

```python
y_hat = model.predict(x_test)
print(f'RMSE {np.sqrt(mean_squared_error(y_test, y_hat)):.3f}')
print(f'R2 Score: {r2_score(y_test, y_hat):.3f}')
```

## GridSearchCV를 이용한 파라미터 튜닝

```python
from sklearn.model_selection import GridSearchCV

params ={'n_estimators':[10, 20, 50, 100, 200, 250],
     'max_features':[2,4,6,8,10],
     'bootstrap':[True, False]}
# max_features: 활용할 최대 독립변수 개수
# bootstrap: 비복원 추출여부

model = RandomForestRegressor()
gs = GridSearchCV(model, params, cv=5, scoring='neg_mean_squared_error')
gs.fit(scaled_x, y)

print('Best Parameters:', gs.best_params_)

# 모델의 학습을 데이터가 들어올때마다 계속 다시 하는 것이 아니라 최적의 파라메터를 파일에 기록/저장해 놓고 학습된 모델을 예측에 사용
```

```python
y_hat = gs.best_estimator_.predict(x_test)
print(f'RMSE {np.sqrt(mean_squared_error(y_test, y_hat)):.3f}')
print(f'R2 Score: {r2_score(y_test, y_hat):.3f}')
```

## 최적화된 모형 저장

```python
import pickle

with open('model.dat', 'wb') as f:
    pickle.dump(gs.best_estimator_, f)
print('모델 저장 완료')
```

```python
import os

if os.path.exists('model.dat'):  # 파일이 존재하는지 여부 확인
    with open('model.dat', 'rb') as f:
        loaded_model = pickle.load(f)  # 저장되어있는 파일 불러오기

y_hat = loaded_model.predict(x_test) # 저장된 모델로 예측하기
print(y_hat[:10])
```

# Bagging - Classifier

## 와인 품종 분류

## 패키지 로딩

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_wine
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

## 데이터 로딩 및 확인

```python
wine = load_wine()
# print(wine.DESCR)
x = wine.data
y = wine.target
df = pd.DataFrame(x, columns=wine.feature_names)
df['class'] =  y
df.head()
print(df['class'].value_counts())
```

## 데이터 스케일링 및 분할

```python
scaled_x = StandardScaler().fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(scaled_x, y, test_size=0.3, stratify=y, random_state=10)
```

## 모델 생성 

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

## 모델 평가

```python
y_hat = model.predict(x_test)
con_mat = confusion_matrix(y_test, y_hat)
print(con_mat)

report = classification_report(y_test, y_hat)
print(report)

print(f'정확도: {accuracy_score(y_test, y_hat):.3f}')
```

## 트리 시각화

### 하이퍼 파라메터 튜닝

```python
from sklearn.model_selection import GridSearchCV

params = {'max_depth':[3,4,5,6,10], 'min_samples_split':[3,6,9,10,20,30]}
# refit=True: 가장 좋은 파라메터 설정으로 재학습한 후 모델을 best_estimator_에 저장 (default)

gs = GridSearchCV(model, params, scoring='accuracy', cv=5)
gs.fit(x, y)
```

```python
print('최적 파라메터:', gs.best_params_)
print(f'최적 파라메터 정확도" {gs.best_score_:.3f}')  
```

### Feature Importance
- 의사결정 나무 알고리즘이 학습을 통해 규칙을 정하는데 있어서 피처의 중요한 역할지표를 제공하는 속성
- 일반적으로 값이 높을수록 해당 피처의 중요도가 높다는 의미로 해

```python
im = gs.best_estimator_.feature_importances_
# 각 독립변수의 중요도가 넘파이 다차열 배열로 반환
```

```python
importance = {k:v for k,v in zip(wine.feature_names, im)}
df_importance = pd.DataFrame(pd.Series(importance), columns=['importance'])  # 인덱스를 가진 시리즈 객체로 넣기
df_importance = df_importance.sort_values('importance', ascending=False)
df_importance
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(df_importance, x='importance', y=df_importance.index)
# x=중요도 값, y=각 인덱스(독립변수)별
plt.show()
```
