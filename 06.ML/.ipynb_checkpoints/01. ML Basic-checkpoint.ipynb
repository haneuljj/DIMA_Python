# Cost Function 그리기

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cost Function
# w: 가중치 (초기:임의값)
# x: 독립변수
# y: 종속변수
# b: 절편 (초기: 임의값)

def MSE(w, x, y, b):
    s = 0
    for i in range(len(x)):
        s += (y[i] - (w*x[i]+b))**2   # (예측 정답 - 예측값(회귀식을 통한)), s: 오차값
        
    return s / len(x)

# y = 2x + 0.1 모델 가정 
x = [1., 2., 3.]   # 독립변수, feature
y = [2.1, 4.1, 6.1]  # 종속변수, label
b = 0.1


w_val = []   # w값을 저장할 리스트
cost_val = []  # 오차값을 저장할 리스트

for w in np.linspace(-4, 8, 20):
    c = MSE(w, x, y, b)
    w_val.append(w)
    cost_val.append(c)

#### cost값을 이용한 그래프 그리기
plt.plot(w_val, cost_val)
plt.xlabel('Weight')
plt.ylabel('Cost')
plt.show()

# weight가 2일 때 손실이 최소, 최적화 상태
```

# 가중치 학습

```python
from sklearn.datasets import load_diabetes   # load_dataset name
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

diabetes = load_diabetes()   # 당뇨병 환자 데이터 로드, bunch객체로 제공(딕셔너리와 비슷)
# print(diabetes.DESCR)

df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)  
# bunch객체.data -> data를 넘파이 다차원배열로, diabetes.feature_names -> 독립변수명들
#df.head()

# age, sex 컬럼이 이미 scaling이 되어있어서 수치로 변환된 데이터셋 보여줌
df['target'] = diabetes.target   # 데이터의 종속변수 가져옴 
#df.head()

w = 1.0   # 초기 가중치(임의값)
b = 1.0   # 초기 절편(임의값)

print(np.min(df['bmi']), np.max(df['bmi']))   # 시작점과 끝점값 선정 이유? bmi값의 최솟값이 대략 -0.1, 최대값이 0.17
pt1 = (-0.1, -0.1*w+b)   # 회귀선의 시작점으로 사용될 x와 y값
pt2 = (0.16, 0.16*w+b)   # 회귀선의 끝점으로 사용될 x와 y값
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]], 'r')   # plot([x좌표 리스트], [y좌표 리스트]) # 초기의 기울기 적용했을 때 선형회귀선

plt.scatter(df['bmi'], df['target'])   # bmi와 종속변수y의 상관관계
plt.xlabel('BMI')
plt.ylabel('y')
plt.show() 
```

```python
for i in range(100):  # 100번 반복 학습 
    for x_i, y_i in zip(df['bmi'], df['target']):    # x = bmi컬럼의 값, y = target컬럼의 값 
        y_hat = w*x_i + b
        w = w - (y_hat - y_i) * x_i   # 비용함수를 가중치에 대해 미분한 값
        b = b - (y_hat - y_i)  # 비용함수를 절편에 대해 미분한 값

pt1 = (-0.1, -0.1*w+b)
pt2 = (0.16, 0.16*w+b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]], 'r')
plt.scatter(df['bmi'], df['target'])  
plt.xlabel('BMI')
plt.ylabel('y')
plt.show() 
```

# 단변량 데이터의 Linear regression
- 사이킷런을 이용함
- https://scikit-learn.org

```python
from sklearn.linear_model import LinearRegression   # scikit의 선형회귀모형을 다루는 클래스
import numpy as np

# 독립변수가 스칼라값이 아닌 다차원 배열로 들어가야 함  -> [[1]]
# 1) 데이터 준비
x = np.array([1, 3, 2, 4, 7, 4, 9, 2, 3, 2, 6, 3, 2, 7])  # (14,) -> (14,1)
x = np.expand_dims(x, axis=1)  # x = x.reshape[-1,1]
y = np.array([3, 9, 6, 7, 10, 6, 12, 2, 4, 3, 8, 5, 3, 10])

# 2) 모델 준비
model = LinearRegression()

# 3) 학습(fitting)
model.fit(x, y)   # y = wx + b 라는 선형식을 만드는 것(최적의 w, b를 찾는 것)

# 4) 평가
r_square = model.score(x, y)
print('R squared:', r_square)

# 5) 예측(추론)
x_new = [[7]]  # 새로 들어온 값(독립변수 -> 2차원)
y_hat = model.predict(x_new)   # 예측값(y_hat)
print('예측값:', y_hat)
print('정답: 10')
```

# Linear Regression - 2

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split  # 학습데이터와 평가데이터를 지정한 비율로 분리시켜주는 함수
import matplotlib.pyplot as plt
import numpy as np

# 1) 데이터 준비
np.random.seed(0)
x = np.random.rand(100, 1) * 10    # 0이상 10미만 범위의 100 x 1의 행렬 생성
y = (x * 2.3) + np.random.rand(100, 1) * 5.4

plt.plot(x, y, 'o')
plt.show()

# 2) 모델 준비
model = LinearRegression()

# 3) 학습데이터와 평가데이터 분리
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, #train_size=0.7,
                                                   random_state = 10) # 분할로 가져오는 랜덤 데이터를 고정, 동일한 데이터로 분할되도록

# 4) 학습
model.fit(x_train, y_train)

# 5) 평가
r_square = model.score(x_test, y_test)
print(f'결정계수: {r_square:.2f}')
print('추정계수(가중치):', model.coef_)
print('절편:', model.intercept_)   # x와 y의 관계로 설명할 수 없는 오차

# 6) 예측 및 시각화
plt.plot(x, y, 'o')
plt.plot(x, model.predict(x), 'r')  # 예측값으로 만들어진 선
plt.show()
```

# [문제] 키와 몸무게를 이용해 모델을 학습 시킨 후, 키 170인 사람의 몸무게를 예측하는 프로그램을 작성하세요

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score   # MAE, MSE, R squared 평가지표
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 1) 데이터 준비
df = pd.read_csv('./dataset/body.csv')
x = df['Height'].values.reshape(-1, 1)   # .values: series의 값이 1차원 벡터값으로,  reshape(-1, 1): 2차원 배열로
# x = df[['Height']]
y = df['Weight']

# 학습데이터와 평가데이터 분리
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=10)

# 2) 모델 준비
model = LinearRegression()

# 3) 학습
model.fit(x_train, y_train)

# 4) 평가
#r_square = model.score(x, y)
#print('R squared:', r_square)
y_hat = model.predict(x_test)   # 평가 데이터를 예측에 사용하여 예측값 구하기
print(f'MAE: {mean_absolute_error(y_test, y_hat):.3f}')   #평가지표(정답, 예측값)
print(f'RMSE: {np.sqrt(mean_squared_error(y_test, y_hat)):.3f}') # RMSE: MSE에 루트 적용
print(f'r2 score: {r2_score(y_test, y_hat):.3f}')

# 5) 예측(추론)
x_new = [[170]]  
y_hat = model.predict(x_new)   
print('키 170.0인 사람의 예측 몸무게:', y_hat[0])

# 6) 시각화
#plt.scatter(x, y)
plt.plot(x, y, 'o')
plt.plot(x, model.predict(x), 'r')
plt.show()
```

```python
# 키가 169 ~171 인 사람들의 평균 몸무게
df[df['Height'].between(169,171)]['Weight'].mean()   # series.between(시작범위, 끝범위):특정 범위에 있는 값 가져오기 --> T/F로 반환
```

# [실습] 캘리포니아 집값 예측
- 1990년 캘리포니아 각 블록 그룹의 주택 가격 데이터

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1) 데이터 준비
data = pd.read_csv('./dataset/california_housing_train.csv')
print(data.shape)
display(data.head())
x = data[data.columns[2:-1]]  # 위도 경도 컬럼 제외, 타겟 컬럼 제외
y = data['median_house_value']

# 2) 모델 준비
# model = LinearRegression()
model = RandomForestRegressor()

# 3) 학습
model.fit(x, y)

# 4) 평가
test_data = pd.read_csv('./dataset/california_housing_test.csv')
print(test_data.shape)
test_x = test_data[test_data.columns[2:-1]]
test_y = test_data['median_house_value']
print('R square:', model.score(test_x, test_y))

# 5) 예측
predict_data = test_x[10:11]
y_hat = model.predict(predict_data)
print('예측값:', y_hat)
print('실제값:', test_y[10])

# 6) 시각화
y_hat = model.predict(test_x)
plt.figure(figsize=(10,5))  # 그래프 사이즈 지정
# 확률밀도그래프: 히스토그램과 유사하나 각 데이터의 구간별 빈도수를 확률적으로 추정하여 부드러운 곡선으로 나타낸다
sns.kdeplot(test_y, label='y', fill=True)   
sns.kdeplot(y_hat, label='y_hat', fill=True)
plt.legend()  # 범례표시
plt.show()
```
