# [실습] yes24에서 파이썬 도서 검색하기
<pre>
    - yes24 사이트에서 도서 검색 후 평점이 9.6점 이상인 도서 제목과 가격, 평점 가져오기
</pre>

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup


driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=webdriver.ChromeOptions())
driver.get('https://www.yes24.com/main/default.aspx')
ele = driver.find_element(By.ID, 'query')  # 검색창 태그 찾기
ele.send_keys('파이썬')   # 검색창에 파이썬 입력
ele.submit()
time.sleep(1)

driver.find_element(By.CSS_SELECTOR, 'option[value="120"]').click()   # 셀렉터 박스에 '120개' 선택, onchange option
```

```python
soup = BeautifulSoup(driver.page_source, 'lxml')

book_list = soup.find('ul', attrs={'class','sGLi tp_book tp_chkG tp_list'})  # 책 전체 리스트 태그 선택
books = book_list.find_all('li')   # 각 책 별 리스트 가져오기
print(len(books))

cnt = 1
for b in books:
    try:
        book_ratings = b.select_one('#yesSchList > li > div > div.item_info > div.info_row.info_rating > span.rating_grade > em').text  # 별점 태그 가져오기, 별점 없는 책 건너뛰어서 예외 처리
    except:
        continue

    if float(book_ratings) >= 9.6:   # 평점 9.6점 이상이면
        b_t = b.find('div', attrs={'class':'info_row info_name'})
        book_title = b_t.find('a', attrs={'class':'gd_name'}).text  # 책 제목 태그
        
        b_p = b.find('strong', attrs={'class':'txt_num'})
        book_price = b_p.find('em', attrs={'class':'yes_b'}).text   # 책 가격 태그
        
        print(f'[{cnt}] 제목: {book_title} | 가격: {book_price}원 | 평점: {book_ratings}점')
        cnt += 1
    
```

```python
soup = BeautifulSoup(driver.page_source, 'lxml')

book_list = soup.find('ul', attrs={'class','sGLi tp_book tp_chkG tp_list'})
books = book_list.find_all('li')
print(len(books))


for page in range(1,11):
    print(f'------- page: {page} -------')
    for book in books:        
        title = book.find('a', attrs={'class':'gd_name'})
        # <em class='yes_b'> 태그가 도서 가격, 평점 등 다른 곳에서도 사용되어 상위태그 먼저 접근 필요
        price = book.find('strong', attrs={'class':'txt_num'})
        rating = book.find('span', attrs={'class':'rating_grade'})
        if not rating: continue  # 평점이 없는 도서는 pass
    
        rating = rating.find('em', attrs={'class':'yes_b'}).text
        if float(rating) < 9.6: continue  # 평점이 9.6 미만이면 pass
    
        print(f'{title.text} | {price.text} | {rating}')
    
        
    time.sleep(5)
    el = f'#goodsListWrap > div.sGoodsPagen > div > a:nth-child({page+1})'
    print(el)
    driver.find_element(By.CSS_SELECTOR, el).click()
```

## iframe을 사용한 웹 페이지 스크래핑하기
- iframe: 다른 html 페이지를 현재 페이지에 중첩
- iframe 의 내용을 크롤링하려면 포커스를 옮겨서 크롤링 해야함

- iframe 태그의 id값을 찾는다.
    - ex) frame = driver.find_element(By.ID, 'entryIFrame')
- 해당 iframe으로 전환한다.
    - ex) driver.switch_to.frame(frame)
- 이후 iframe 내의 element를 검색하여 데이터를 스크래핑 한다.
- iframe 내의 스크래핑 작업이 끝난 후 기본 프래임의 내용을 스크래핑 하기 위해 기본프레임으로 전환한다.
    - ex) driver.switch_to.default_content()
