## 스타일

```python
import numpy as np

plt.plot(np.arange(6), np.arange(6)**2,
        color='red',
        linestyle='dashdot',    # linestyle: solid, dashed, dotted, dashdot
        linewidth=2.5,
        marker='o',
        markeredgecolor='green',
        markerfacecolor='yellow',
         markersize=9.3
        )   
plt.show()
```

- 상징을 사용해서 문자열 형태로 스타일을 설정할 수도 있다.


```python

import numpy as np

x = np.linspace(0, 2*np.pi, 100)  # 0에서 2파이값을 10개로 균등분할

# 적색 선, 원형 마커, 실선 / 녹색 선, 삼각형 마커, 실점선 / 청색 선, x 마커, 점선
plt.plot(x, np.sin(x), 'ro--', x, np.cos(x), 'g^-.', x, -np.sin(x), 'bx:')
plt.xlim(0, 2*np.pi)  # x축의 범위 지정
plt.show()
```

```python
t = np.arange(0., 5., 0.2)   # range와 기능은 같으나 정수형뿐만 아닌 실수형도 가능

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^:')
plt.show()
```

# 범례 표시하기
- plot() 함수에 label 키워드 이나를 설정함으로써 그래프의 이름을 지정할 수 있다.
- 이렇게 지정된 이름은 legend() 함수를 통해 범례로써 표시할 수 있다.
- loc 파라미터에 튜플로 두 개의 값을 지정하면 해당 위치에 범례가 표시된다.
<pre>
    'best': 그래프에서 자동으로 최적의 위치에 표시해줌
    'upper right', 'upper left', 'lower right', 'lower left', 'right', 'center right', 'center left', 'upper center', 'lower center', 'center'
</pre>

```python
a = np.arange(0, 3, 0.02)
c = np.exp(a)
d = c[::-1]

plt.plot(a, c, 'r--', label = 'Model Length')
plt.plot(a, d, 'b:', label = 'Data Length')
plt.plot(a, c+d, 'g', label = 'Total Message Length')

plt.legend(loc='upper center', shadow=True)  # 범례 지정
plt.show()
```

# 서브플롯 그리기
- subplot() 함수를 이용해서 하나의 figure 에 여러 개의 그래프를 그릴 수 있다.
- 주요 인자: nrows, ncols, index를 지정하는 3개의 정수를 이용하여 서브플롯의 위치를 지정할 수 있다.

```python
x1 = np.linspace(0.0, 5.0)  # 분할할 데이터의 수를 입력하지 않으면 기본 값은 50이다.
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1) # nrows =2 , ncols = 1, index = 1
plt.plot(x1, y1, 'o-')
plt.title('1st graph')
plt.xlabel('time (s)')
plt.ylabel('Damped')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.title('2nd graph')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()  # 서브플롯의 레이아웃을 자동으로 맞춰준다
plt.show()
```

- subplots() 함수를 이용해서 서브플롯을 그릴 수도 있다.
- subplots() 함수는 Figure와 Axes 객체를 반환한다.
  - Axes 클래스: 공간이 확보된 이미지의 영역정보를 가지고 있다.
- 주요 인자
  - nrows, ncols: 서브플롯의 행, 열값 지정
  - sharex, sharey: boolean 값  (default:False)  x, y 축 사이의 공유제어 속성

```python
fig, axes = plt.subplots(2, 2)#, sharey=True)  # y축값 공유
fig.set_size_inches((16, 10))  # Figure의 크기를 설정

axes[0][0].scatter(np.random.random(100), np.random.random(100), color='cyan', edgecolor='red', alpha=0.5)
axes[0][1].scatter(np.random.random(100), np.random.random(100), color='magenta', edgecolor='green', alpha=0.5)
axes[1][0].scatter(np.random.random(100), np.random.random(100), color='yellow', edgecolor='blue', alpha=0.5)
axes[1][1].scatter(np.random.random(100), np.random.random(100), color='white', edgecolor='black', alpha=0.5)

axes[0][0].set_title('subplot0')
axes[0][1].set_title('subplot1')
axes[1][0].set_title('subplot2')
axes[1][1].set_title('subplot3')

plt.show()
```

# 막대그래프 그리기
- bar() 함수를 이용해서 막대 그래프를 그릴 수 있다

```python
x1 = np.arange(0, 20)
y1 = x1 * 2

x_ticks = np.arange(len(x1))

plt.bar(x1, y1, label='Blue bar', color='blue', edgecolor='yellow', linewidth=3)
plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('Bar Chart Example')
plt.xticks(x_ticks, x_ticks+1)
plt.legend()
plt.show()
```

- barh() 함수를 이용해서 수평 막대 그래프를 그릴 수 있다.

```python
x1 = np.arange(0, 20)
y1 = x1 * 2

y_ticks = np.arange(len(x1))

# 수평 막대 그래프를 그린다고해서 x와 y의 위치가 바뀌는 건 아님
plt.barh(x1, y1, label='Blue bar', color='blue', edgecolor='yellow', linewidth=3)
plt.xlabel('bar height')
plt.ylabel('bar number')
plt.title('Bar Chart Example')
plt.yticks(y_ticks, y_ticks+1)
plt.legend()
plt.show()
```

```python
# 하나의 눈금에 두 개의 막대그래프 그리기
import numpy as np
import matplotlib.pyplot as plt

labels = ['Seoul', 'Busan', 'Incheon', 'Gwangju', 'Jeju']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x_ticks = np.arange(len(men_means))
width = 0.35  # 막대 그래프의 폭 크기
plt.bar(x_ticks-width/2, men_means, width=width, color='dodgerblue', label='Men')
plt.bar(x_ticks+width/2, women_means, width=width, color='violet', label='Women')

plt.xlabel('City')
plt.ylabel('Score')
plt.xticks(x_ticks, labels)
plt.title('Score by group and gender')
plt.legend()
plt.show()
```

# 산점도 그리기
- 산점도(scatter plot)은 직교 좌표계를 이용해서 두 변수 간의 상관관계를 나타내는 plot이다.
- scatter() 함수를 이용해서 산점도를 그릴 수 있다.

```python
np.random.seed(0)
for color in ['blue', 'orange', 'green']:
    n = 50
    x, y = np.random.rand(2, n)  # [0, 1) 0이상 1미만의 난수를 2 x 50 모양으로 생성 (2차원)
    scale = 190 * np.random.rand(n) + 10  # [10, 200) 사이의 난수 50개 생성
    plt.scatter(x, y, color=color, s=scale, label=color, alpha=0.3, edgecolor='none')  # none; 테두리가 없는 마커

plt.legend()   # loc 디폴트는 'best'
plt.grid(True)   # 격자무늬 생성
plt.show()
```

# 히스토그램 그리기
- hist() 함수를 이용해 히스토그램을 그릴 수 있다.
- 자료의 분포를 몇 개의 구간으로 나눈 후에 각 구간에 해당하는 데이터의 빈도수를 표로 나타낸 것을 도수분포표(frequency table)라고 한다.
- 히스토그램은 도수분포표에서 구간별 빈도수를 막대그래프로 나타낸 것이다.

```python
y = np.random.randn(1000)   # 평균이 0 , 표준편차는 1인 표준정규분포로부터 난수 1000개 생성
x = np.arange(len(y))
plt.bar(x, y)
plt.title('Raw Data <Bar Chart>')
plt.show()

plt.hist(y, bins= 50)  # bins=50, 표현할 데이터를 50개의 구간으로 나눈다.
plt.title('Histogram')
plt.show()

plt.hist(y, cumulative=True, bins=20)  # 누적 도수분포표
plt.title('Cumulative Histogram')
plt.show()
```

# 원 그래프 그리기
- pie() 함수를 이용해서 원 그래프를 그릴 수 있다.

```python
labels = ['python', 'java', 'c']
# autopct: wedge위에 표시할 레이블 (default: None)
# pie()함수에 전달하는 값은 굳이 비율 값을 전달하지 않고 일반적인 값을 전달하더라도 값들의 총합에 대한 비율을 알아서 계산해서 표시함
plt.pie([0.3, 0.35, 0.4], labels=labels, autopct='%.1f%%')  # [파이 그래프에 들어갈 비율]
plt.title('Popular Language Ranking')
plt.show()
```

```python
sections = [0.3, 0.35, 0.4]
labels = ['python', 'java', 'c']
colors = ['coral', 'dodgerblue', 'yellowgreen']

plt.pie(
    sections,  # 조각의 크기를 설정하는 기본 값
    explode = (0.1, 0, 0),  # 조각을 중심으로부터 얼마나 떨어지게 할 것인지 설정한다, (0.1: 반지름의 10%만큼 벗어남)
    colors=colors,   # 조각의 색깔을 결정
    labels=labels,    # 조각의 이름을 표시
    autopct='%.2f%%',  # 비율을 표시하는 문자열 형식 지정
    shadow=True,
    startangle = 90,  # 원형의 시작 부분을 각도로 표시. (기본값은 3시방향, 0도)
    counterclock=False,  # 반시계 방향으로 표시할지를 결정 (기본값은 True)
    rotatelabels=True   #  조각의 이름을 회전시킨다 (기본값은 False)
)
plt.title('Popular Language Ranking')
plt.show()
```

# boxplot 그리기
- boxplot은 데이터 집합의 범위와 중앙값 등 통계 수치를 확인하고 데이터의 이상치를 확인하고자 하는 목적으로 사용한다
- 최소값: 제 1사분위에서 1.5IQR(interquartile range, Q3-Q1)을 뺀 위치
- 제 1사분위(Q1): 중앙값 기준으로 하위 50% 중의 중앙값, 전체 데이터 중 하위 25%에 해당하는 값
- 제 2사분위(Q2): 중앙값, 전체 데이터 중 50%에 위치
- 제 3사분위(Q3): 중앙값 기준으로 상위 50% 중의 중앙값, 전체 데이터 중 상위 25%에 해당하는 값
- 최대값: 제3사분위에서 1.5IQR을 더한 위치
- IQR(사분위수 범위) = Q3-Q1
- 최대값, 최소값 범위를 벗어나는 값을 이상치 데이터로 간주한다.

```python
np.random.seed(0)

data_a = np.random.normal(0, 2.0, 1000)  # 평균은 0, 표준편차는 2.0, 1000개
data_b = np.random.normal(-3, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

plt.boxplot([data_a, data_b, data_c])   # 그래프를 그릴 객체들을 리스트로 전달
plt.ylim(-10, 10)
plt.xlabel('Data Type')
plt.ylabel('Value')

plt.show()
```
